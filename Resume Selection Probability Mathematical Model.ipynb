## Model Explanation

1. The model is based on the Logistic Model.
2. $x$ is the vector of variables.
3. $f(x)$ is the probability of resume selection.

5. All the 4 coefficients in our equation range around 1.
    - as the current macroeconomic outlook for India is positive and getting better by months and years, this coefficient should be more than 1 such as 1.1 or 1.25.
    - as my ug degree is not from a particularly reputated university, the coefficient could be 0.8
    - as my percentages in ug degree is 66 let the coefficient be 0.66 for simplicity
    - for the coefficient of portfolio projects, a survey and data collection would be required to assess the level of portfolio projects a typical candidate presents, so I wouldn't comment much on this, but it still ranges around 1. If my projects are better than others then it could be 1.1 for instance otherwise it could be 0.9 or 0.85, for example.

<br>

6. The variables here should measure expertise in that subject, so its domain can be between 1 and 10 or we can take it as any real number, if we take an ordinal perspective, assuming a higher number with higher skills but no measure of how much higher is one number to another, just simply higher.
7. Many variables have correlations with each other and many of them can also exist independent of others
    - for instance in our first pair of correlated variables we have stats and prob, where some stats can be known without knowing prob and prob can also be known without stats but if one knows them together then they both complement each other, therefore, in our model stats increases further if prob increases and prob also increases on its own without any knowledge of stats
    - many other variables in this model are modelled in the same way as explained in the previous line
    - data analysis and related skills have correlation with Python, R, Jupyter, SQL, PowerBI, Excel, Tableau and other things,but for simplicty I have assumed in the model that it has no correlation with other variables except for Time Series Analysis and Feature Engineering. Without this assumption we would have to show this correlation with all those variables and the equation would get double in size.
 
 <br>
 
8. Java is mostly used for OOP related work, so we assume in the model that OOP has correlation only with Java and not with Python
    - OOP also has its own marginal contribution
    - for the same thing, Calculus has its own marginal contribution, but you can't know nonlinear optimization without knowing Calculus, so in our model knowledge of nonlinear optimization increases Calculus's contribution but when calculus is zero the other is already zero.
    
9. We can futher optimize the relationship in this model but adjusting the exponents of variables.
    - for instance if it is a finance analyst job that I am applying for then the contribution of finance and economics variables should increase and for that we can increase their exponents, on the other hand, the exponents on Java should decrease as it is least related thing in Finance Analysis or we can even take logarithm of the Java variable.
    - for simplicity we have taken all exponents to be 1.
    - a group variables can also have an exponent, that will increases or decreases that group's marginal contribution depending on the job applied for.
    - for example, for the previous line, if I am applying for financial analyst job then the finance and economics group can have its exponent set to more than 1, how much more than 1 will depend on how deeply that group affects the probability of resume selection.
    - in the above equation lower exponents suggest that having some skills of that variable greatly enhances the chance of selection while higher skills in that variable won't much increase the chance of selection. For the higher exponents, they suggest that a relatively higher skill level of that variable is required for this job and so knowing basics won't greatly enhance the chance of selection but as much expert you become in that variable the chance improves as much rapidly.
    - so exponents also, don't have any cardinal characteristics in this model, they are majorly denoting the ordinal nature of the model.
    - when the exponent is 1, the chance increases linearly with skill level in that variable, that means, for each unit increase in skills in that variable, the chance will by a fixed certain units

**As there are dozens of variables explaining each one would be a clutter, so I have given a broader overview of the model**

- For a general overview, there are only three kinds of variables:
    - ones having only their own marginal contribution
    - ones having correlation with other variables and their own marginal contribution
    - ones having correlation with other variables and no contribution of their own without the other variables that they correlate with.
    
**Working of the model**

- Variables that increases the skillset increases the probability towards 1.
- Variables that decrease the skillsest or coefficients that scale down the skillset decreases the probability.
